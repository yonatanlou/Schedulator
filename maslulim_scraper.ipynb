## Gathering all the maslulim from all the facaulties.

```python
def hugim_scraper(faculties):
    """ return  a dictionay whcih contains dictionay = {faculty_name:{hug_id: hug_name}}
    faculties = dict of faculty_name:faculty_id
    """
    faculty_dict = dict()
    browser = webdriver.Chrome()
    browser.get("https://shnaton.huji.ac.il/")
    
    #first loop over faculty
    for fac_name, fac_id in faculties.items():
        select = Select(browser.find_element_by_id('faculty'))
        select.select_by_value(f"{fac_id}") #Pass value as string
        sleep(0.5)
        hugim_options = browser.find_elements(By.XPATH,'//*[@name="hug"]/option')
        #second loop over hugim
        hugim_dict = dict()
        for hug in hugim_options:
            if hug.get_attribute("value") == "0" or hug.text == "בחרו חוג":
                pass
            else:
                hugim_dict[hug.text] = hug.get_attribute("value") 
        faculty_dict[fac_name] = hugim_dict
    return faculty_dict

all_hugim = hugim_scraper(faculties)
```

```python
def hugim_printer(all_faculties_dict):
    for fac_name ,hug_dict in all_faculties_dict.items():
        print(fac_name)
        for hug_name, hug_id in hug_dict.items():
            print(hug_name, hug_id, sep=": ")
        print("\n")
hugim_printer(all_hugim)
```
